# Scraping 2017 French presidential election (second round).

```python
#import libraries
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# Create an URL object
Url1 = 'https://www.interieur.gouv.fr/Elections/Les-resultats/Presidentielles/elecresult__presidentielle-2017/(path)/presidentielle-2017/index.html'
# Create object page
page = requests.get(Url1)
```

```python
page
```

```python
page
soup = bs(page.text, 'lxml')
soup
```

```python
table2=soup.find('select', id='listeDpt')
index=[]
```

```python
#for the index
for j in table2.find_all('option')[1:]:
    index.append(j.text)
```

```python
index
```

```python
mydata=pd.DataFrame(columns=["Macron","Le Pen"],index=index)
```

```python
#scraping multiple pages
for url in range(0,107):
    req = requests.get(urls[url])
    soup = bs(req.text, 'html.parser')
  
    table1 = soup.find('table', class_='table table-bordered tableau-resultats-listes')
    for j in table1.find_all('tr')[1:]:
        for i in j.find_all('td',style='text-align:right'):
            row.append(i.text)
    length = len(mydata)
    mydata.loc[length] = row
    row=[]
```
